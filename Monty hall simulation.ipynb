```python
import matplotlib.pyplot as plt
import random
from matplotlib.pyplot import figure
doors=['1','2','3']
prizes=['goat','goat','car']
def assign(n,l):
    s=[]
    for i in range (len(l)):
        s.append([n[i],l[i]])
    return s
def check_result(new_dor):
    if new_dor[1]=='car':
        return True
    else:
        return False
def switcher(s,l,choice):
    if s=='y':
        door=l[choice-1]
    else:
        door=l[choice-2]
    return door
count=0
switch='n'
countt=[]
cpt=0
x=100000
l=[]
cptt=[]
for i in range(1,x):
    tuble=assign(doors,prizes)
    n=random.randint(0,2)
    for i in range (len(tuble)):
        if tuble[i][1]!='car' and i!=n:
            tuble.remove(tuble[i])
            break
    door=switcher(switch,tuble,n)
    if check_result(door)==True:
        count+=1
    cpt+=1
    cptt.append(cpt)
    countt.append(count/cpt)
v=count/x
print('\nWin probaility: '+str(v)+'              Loss probability: '+str((1-v)))
figure(num=None, figsize=(18, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(cptt,countt);
```
